# 贪心算法
- seq2seq中，我们使用贪心搜索来预测序列
  - 将当前时刻预测最大的词输出
- 但贪心很可能不是最优的，效率最高的

# 穷举搜索
- 最优算法：对所有可能的序列，计算他的概率，然后选取最好的那个
- 如果输出字典大小为n，序列最长为T，那么我们需要考察 n_T  个许立群
  - n=10000, T=10: n_T = 10_40
  - 计算上不可行
